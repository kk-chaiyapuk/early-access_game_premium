## Google Play Search

```python
result = googplay.search(
        "Action",
        lang="en",  # defaults to 'en'
        country="us",  # defaults to 'us'
        n_hits=50)  # defaults to 30 (= Google's maximum))

result = pd.DataFrame(result)
```

```python
result.columns
```

```python
result
```

## Google Play Game Detail

```python
all_mobile_game = []
for app_id in result['appId']:

    app = googplay.app(app_id,
                      lang='en', # defaults to 'en'
                      country='us') # defaults to 'us'
    all_mobile_game.append(app)

all_mobile_game = pd.DataFrame(all_mobile_game)
```

```python
marvel_snap_review_all = pd.DataFrame(googplay.reviews_all('com.nvsgames.snap',
    sleep_milliseconds=0, # defaults to 0
    lang='en', # defaults to 'en'
    country='us', # defaults to 'us'
    # sort=Sort.MOST_RELEVANT, # defaults to Sort.MOST_RELEVANT
    filter_score_with=None # defaults to None(means all score)
))
```

```python
top_worst_review = marvel_snap_review_all.query("score==1").sort_values("thumbsUpCount", ascending=False)["content"].head(5)

for review in top_worst_review:
    print(review, end="\n\n")
```
